```python
#Importing important libraries
import pandas as pd
from sklearn.preprocessing import normalize
```

```python
#Loading the dataset
patients_data = pd.read_csv("plan b data.csv", index_col=0)
```

```python
#In practice, feature selection should be done after data pre-processing, so ideally, all the categorical variables are encoded into numbers
numerics = ['int16','int32','int64','float16','float32','float64']
numerical_vars = list(patients_data.select_dtypes(include=numerics).columns)
patients_data = patients_data[numerical_vars]
patients_data
```

```python
#data type of variables
patients_data.dtypes
```

```python
patients_data.isnull().sum()
```

```python
#Creating the traning set of the features
data = patients_data.drop(['Diagnosis'], axis=1)
```

```python
normalize = normalize(data)
```

```python
data_scaled = pd.DataFrame(normalize)
data_scaled.var()
```

```python
#storing the variance and name of variables
variance = data_scaled.var()
columns = data.columns
```

```python
#saving the names of variables having variance more than a threshold value

variable = [ ]

for i in range(0,len(variance)):
    if variance[i]>=0.006: #setting the threshold as 1%
        variable.append(columns[i])
```

```python
variable
```

```python
# creating a new dataframe using the above variables
new_data = data[variable]

# first five rows of the new data
new_data.head()
```
