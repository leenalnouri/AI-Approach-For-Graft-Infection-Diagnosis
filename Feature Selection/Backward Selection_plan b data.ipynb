```python
#importing relevant libraries
import pandas as pd
pd.options.mode.chained_assignment = None
import pylab as pl
import numpy as np
import scipy.optimize as opt
from sklearn import preprocessing
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
#Reading the data
patients_data = pd.read_csv("plan b data.csv", index_col=0)
```

```python
patients_data
```

```python
#In practice, feature selection should be done after data pre-processing, so ideally, all the categorical variables are encoded into numbers
numerics = ['int16','int32','int64','float16','float32','float64']
numerical_vars = list(patients_data.select_dtypes(include=numerics).columns)
patients_data = patients_data[numerical_vars]
patients_data
```

```python
#Checking missing values in the data
patients_data.isnull().sum()
```

```python
#Creating the traning set of the features
X = patients_data.drop(['Diagnosis'], axis=1)
X
```

```python
#Creating the traning set of the target variable
y = np.asarray(patients_data[['Diagnosis']])
y
```

```python
#importing the models
from mlxtend.feature_selection import SequentialFeatureSelector as SBS
from sklearn.linear_model import LinearRegression
# Sequential Backward Selection(sbs)
sbs = SBS(LinearRegression(),
         k_features=11,
         forward=False,
         floating=False,
         cv=0)
```

```python
sbs.fit(X, y)
sbs.k_feature_names_
```

```python
#using the optimal value of k_features argument in both forward selection and backward elimination. In order to find out the optimal number of significant features
sbs1 = SBS(LinearRegression(),
         k_features=(3,50),
         forward=False,
         floating=False,
         cv=0)
sbs1.fit(X, y)
```

```python
from mlxtend.plotting import plot_sequential_feature_selection as plot_sbs
import matplotlib.pyplot as plt
fig1 = plot_sbs(sbs1.get_metric_dict(), kind='std_dev')
plt.title('Sequential Backward Selection (w. StdErr)')
plt.grid()
plt.show()
```

```python
#the last two blocks of code are for optimization purposes to choose the best number of features to use based on comparing the performance of other number of features used. As we can see from the last graph the best performance occurs when the number of features is 25
```

```python
#Using the optimized number of features (25)
#importing the models
from mlxtend.feature_selection import SequentialFeatureSelector as SBS
from sklearn.linear_model import LinearRegression
# Sequential Backward Selection(sbs)
sbs = SBS(LinearRegression(),
         k_features=25,
         forward=False,
         floating=False,
         cv=0)
```

```python
sbs.fit(X, y)
sbs.k_feature_names_
```
